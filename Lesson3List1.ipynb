```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import math
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import *
```

```python
# Task 1
# Координаты вектора АВ
A=[15,10,3]
B=[24,15,9]
AB=math.sqrt((A[0]-B[0])**2+(A[1]-B[1])**2+(A[2]-B[2])**2)
print("Длина вектора АВ: ",round(AB,2))
```

```python
#Task3.1
# Circle
R = 3
t=np.linspace(-5,5,41)
x,y=np.meshgrid(t,t)
F=x**2+y**2-R**2

plt.figure()
plt.contour(x, y, F, [0], linewidths=2,colors='b')
plt.gcf().gca().axis('image')
plt.gcf().gca().grid(True)
plt.show()
```

```python
#Task3.2
# Elipse
a = 5
b = 3 
t=np.linspace(-6,6,50)
x,y=np.meshgrid(t,t)
F=(x/a)**2+(y/b)**2-1

plt.figure()
plt.contour(x, y, F, [0], linewidths=2,colors='g')
plt.gcf().gca().axis('image')
plt.gcf().gca().grid(True)
plt.show()
```

```python
#Task3.3
# Hyperbole
a = 1
b = 2 

t=np.linspace(-6,6,50)
x,y=np.meshgrid(t,t)
F=(x/a)**2-(y/b)**2-1

plt.figure()
plt.contour(x, y, F, [0], linewidths=2,colors='r')
plt.gcf().gca().axis('image')
plt.gcf().gca().grid(True)
plt.show()  
```

```python
#Task 5.1
fig = plt.figure()
ax = Axes3D(fig)
X, Y = np.mgrid[0:10:2, -5:0:4]
Z = 2*X + 3*Y + 5
Z2 = 2*X + 3*Y + 0
ax.plot_surface(X, Y, Z, rstride=1,cstride=1, color='r')
ax.plot_surface(X, Y, Z2, rstride=1,cstride=1, color='b')
plt.show()
```

```python
#Task 5.2
fig = plt.figure()
ax = Axes3D(fig)
X, Y = np.mgrid[-10:10:3, -10:10:3]
Z = X**2/3 + Y**2/4
Z2 = X**2/9 - Y**2/4 - 9
ax.plot_surface(X, Y, Z,rstride=1,cstride=1, color ='b')
ax.plot_surface(X, Y, Z2, rstride=1,cstride=1, color = 'c')
plt.show ()
```
