```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
k=1
while k==1:
    k=float(input('Ведите значение переменной k функции sin(к*х) (k - не должно равняться 1): '))
```

```python
x = np.linspace(-np.pi*3, np.pi*3, 150)
plt.plot(x, np.sin(x), label = 'sin(x)')
plt.plot(x, np.sin(k*x), label = 'sin(k,x)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.legend(loc = 'lower right')
plt.show()
```
