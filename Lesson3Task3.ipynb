```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import *
```

```python
#Task3.1
# Circle
R = 5
x=[]
x1=[]
y=[]
y1=[]

for i in range(1000):
    xi = R*(i+1)/(1000)
    f=math.sqrt(R**2-xi**2)
    x.append(xi)
    x1.append(-xi)
    y.append(f)
    y1.append(-f)
    
plt.plot(x, y, color='green')
plt.plot(x, y1, color='green')
plt.plot(x1, y, color='green')
plt.plot(x1, y1, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```

```python
#Task3.2
# Elipse
a = 3
b = 5 
x=[]
x1=[]
y=[]
y1=[]

for i in range(1000):
    xi = a*(i+1)/(1000)
    f = b*math.sqrt(1-(xi/a)**2)
    x.append(xi)
    x1.append(-xi)
    y.append(f)
    y1.append(-f)
    
plt.plot(x, y, color='green')
plt.plot(x, y1, color='green')
plt.plot(x1, y, color='green')
plt.plot(x1, y1, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```

```python
#Task3.3
# Hyperbole
a = 1
b = 2 
x=[]
x1=[]
y=[]
y1=[]

for i in range(500):
    xi = a+i/500
    f = b*math.sqrt(-1+(xi/a)**2)
    x.append(xi)
    x1.append(-xi)
    y.append(f)
    y1.append(-f)
    
plt.plot(x, y, color='green')
plt.plot(x, y1, color='green')
plt.plot(x1, y, color='green')
plt.plot(x1, y1, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```
