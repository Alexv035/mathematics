```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 10, 30)
plt.plot(x, np.sin(x), label = 'Plot 1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

r = [np.pi, 2*np.pi, 3*np.pi, 100, 500]

a = np.linspace(0.01, 0.02, 30)
plt.plot(a, r[0]/a, color = 'c')
plt.plot(a, r[1]/a, color = 'b')
plt.plot(a, r[2]/a, color = 'g')
plt.plot(a, r[3]*a/a, color = 'r')
plt.plot(a, r[4]*a/a, color = 'r')
plt.xlabel('a')
plt.ylabel('x')
plt.show()

# на втором графике видно что сверху область х ограничена значением 500, а снизу минимальным значением первого 
# графика - pi/0,02 = 50*pi 
```
